```python
import pandas as pd # Load the Python package panda for data manipulation
import numpy as np  # Load the Python package numpy for some math|
url = 'https://raw.githubusercontent.com/ormarketing/248/master/248.csv' # define url as the link....
df = pd.read_csv(url) # read the object url_248, which is a csv file as a dataframe
```

```python
df.head(10) # return the first 10 rows and columns of the data
```

```python
df.tail(10) # return the last 10 rows and columns of the data
```

```python
import statsmodels.api as sm # import package statsmodels as sm, which is a package for ...stats models, whiche include OLS
price1 = df['REGPR1']-df['CUT1'] # define variable price1 as...
price2 = df['REGPR2']-df['CUT2']
price3 = df['REGPR3']-df['CUT3']
df['const'] = 1 # define a vector of 1s in the dataframe
df['Price1'] = price1
df['Price2'] = price2
df['Price3'] = price3
```

```python
y_brand1 = df['UNITS1'] # define variable y the column 'UNITS1' in the dataframe. This is my DEPENDENT Variable
x_brand1 = df[['const','Price1','Price2','Price3','FEAT1','FEAT2','FEAT3','DISP1','DISP2','DISP3']] #define the matrix x with all INDEPENDENT variables
reg_brand1 = sm.OLS(y_brand1,x_brand1)# do the regression
type(reg_brand1)
results = reg_brand1.fit()
type(results)
print(results.summary())
```

## **What is the regression I ran?**

\begin{equation}
Units1=Intercept+\alpha_1 Price 1+\alpha_2 Price 2+ \alpha_3 Price 3+ \alpha_4 Feat_1+\alpha_5 Feat_2+\alpha_6 Feat 3+\alpha_7 Disp1+\alpha_8 Disp2+\alpha_9 Disp3+Errors
\end{equation}

```python
# how to define log
Ln_y_brand1 = np.log(y_brand1) #This taking the log of sales

# do the semi-log estimation main difference is Ln_y_brand1 in the call,
# i.e., reg_ln_brand1 = sm.OLS(Ln_y_brand1,x_brand1) instead of reg_brand1 = sm.OLS(y_brand1,x_brand1).
reg_ln_brand1 = sm.OLS(Ln_y_brand1,x_brand1)#Rest of the code follow by replacing reg_brand1 by reg_ln_brand1
type(reg_ln_brand1)
results = reg_ln_brand1.fit()
type(results)
print(results.summary())
```

# What does this output tell us?

### Model Selection

We want to use the semi-log model because it might provide a higher adjusted $R^2$.

Remark: the $R^2$ represents the percentage of variations of the DV (i.e., y), that is explained by the "model".

This tells us if this model is a better fit to capture the variance in our data while minimizing error. An adjusted $R^2$ .57 tells us that 57% of the variance in the dependent variable (i.e., units) is captured by "the model" (The rest of the variance is captured by the error term)

### Variable Selection

Of all the variables, only FEAT1 is statistically significant. We know this because it has a p-value < .05.

This means that at a $95\%$ condifdence level, we cannot reject the null hypothesis that parameter estimates for Feature 1 is zero, i.e., therefore we conclude it is significant.





### Variable Analysis

Our price (Price1) doesn't significanlty impact sales ***for the price range in our data set***. It is inelastic for this price range. We do not need to do price optimization for this price range ***we can set the price at the max for this range, every time we discount the product, we are losing money that we could have made***

The competitors prices (Price2 and Price3) do not affect our sales either (within the data range). If they change their price, we do not need to react.

These prior two points tell us that we have ***market power***

# Calculate elasticities

```python
# Create the varaiables we need
avg_Price1 = np.round(np.average(df.Price1),2)
print(f'Average Price Brand 1 [{avg_Price1}]\n', end='')
avg_Price2 = np.round(np.average(df.Price2),2)
print(f'Average Price Brand 2 [{avg_Price2}]\n', end='')
avg_Price3 = np.round(np.average(df.Price3),2)
print(f'Average Price Brand 3 [{avg_Price3}]\n', end='')
avg_Units1 = np.round(np.average(df.UNITS1),2)
print(f'Average Sales Brand 1 [{avg_Units1}]\n', end='')
avg_Units2 = np.round(np.average(df.UNITS2),2)
print(f'Average Sales Brand 2 [{avg_Units2}]\n', end='')
avg_Units3 = np.round(np.average(df.UNITS3),2)
print(f'Average Sales Brand 3 [{avg_Units3}]\n', end='')
#The following coefficients come from the simple linear regression since it had higher R^2
coef_brand1_price1 = 0 #because p > .05
coef_brand1_price2 = 0 #because p > .05
coef_brand1_price3 = 0 #because p > .05
#The following coefficients come from the simple linear regression since it had higher R^2
coef_brand2_price1 = 560.4080
coef_brand2_price2 = -831.2171
coef_brand2_price3 = 0 #because p > .05
#The following coefficients come from the *semi-log* linear regression since it had higher R^2
coef_brand3_price1 = 3.9772
coef_brand3_price2 = 2.1340
coef_brand3_price3 = -5.9328
```

```python
#Calculate Brand 1 elasticities
elas_brand1_brand1 = np.round(coef_brand1_price1 * (avg_Price1 / avg_Units1),2)
print(f'Brand 1 price elasticity from Brand 1 price [{elas_brand1_brand1}]\n', end='')
elas_brand1_brand2 = np.round(coef_brand1_price2 * (avg_Price2 / avg_Units1),2)
print(f'Brand 1 price elasticity from Brand 2 price [{elas_brand1_brand2}]\n', end='')
elas_brand1_brand3 = np.round(coef_brand1_price3 * (avg_Price3 / avg_Units1),2)
print(f'Brand 1 price elasticity from Brand 3 price [{elas_brand1_brand3}]\n', end='')
```

```python
#Calculate Brand 2 elasticities
elas_brand2_brand1 = np.round(coef_brand2_price1 * (avg_Price1 / avg_Units2),2)
print(f'Brand 2 price elasticity from Brand 1 price [{elas_brand2_brand1}]\n', end='')
elas_brand2_brand2 = np.round(coef_brand2_price2 * (avg_Price2 / avg_Units2),2)
print(f'Brand 2 price elasticity from Brand 2 price [{elas_brand2_brand2}]\n', end='')
elas_brand2_brand3 = np.round(coef_brand2_price3 * (avg_Price3 / avg_Units2),2)
print(f'Brand 2 price elasticity from Brand 3 price [{elas_brand2_brand3}]\n', end='')
```

```python
#Calculate Brand 3 elasticities
elas_brand3_brand1 = np.round(coef_brand3_price1 * avg_Price1,2)
print(f'Brand 3 price elasticity from Brand 1 price [{elas_brand3_brand1}]\n', end='')
elas_brand3_brand2 = np.round(coef_brand3_price2 * avg_Price2,2)
print(f'Brand 3 price elasticity from Brand 2 price [{elas_brand3_brand2}]\n', end='')
elas_brand3_brand3 = np.round(coef_brand3_price3 * avg_Price3,2)
print(f'Brand 3 price elasticity from Brand 3 price [{elas_brand3_brand3}]\n', end='')
```
