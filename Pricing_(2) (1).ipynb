```python
import pandas as pd # Load the Python package panda for data manipulation
import numpy as np  # Load the Python package numpy for some math|
# Load data from local file (project 248.csv or Downloads)
try:
    df = pd.read_csv('248.csv')  # project directory
except FileNotFoundError:
    print("File not found")
```

```python
df.tail(10) # return the last 10 rows and columns of the data
```

```python
import statsmodels.api as sm # import package statsmodels as sm, which is a package for ...stats models, whiche include OLS
price1 = df['REGPR1']-df['CUT1'] # define variable price1 as...
price2 = df['REGPR2']-df['CUT2']
price3 = df['REGPR3']-df['CUT3']
price4 = df['REGPR4']-df['CUT4']
df['const'] = 1 # define a vector of 1s in the dataframe
df['Price1'] = price1
df['Price2'] = price2
df['Price3'] = price3
df['Price4'] = price4
# Log-price columns for Log-Log models (all brands)
df['log_Price1'] = np.log(df['Price1'])
df['log_Price2'] = np.log(df['Price2'])
df['log_Price3'] = np.log(df['Price3'])
df['log_Price4'] = np.log(df['Price4'])
```

```python
y_brand1 = df['UNITS1'] # define variable y the column 'UNITS1' in the dataframe. This is my DEPENDENT Variable
x_brand1 = df[['const','Price1','Price2','Price3','Price4','FEAT1','FEAT2','FEAT3','FEAT4','DISP1','DISP2','DISP3','DISP4']] #define the matrix x with all INDEPENDENT variables
reg_brand1 = sm.OLS(y_brand1,x_brand1)# do the regression
type(reg_brand1)
results_brand1 = reg_brand1.fit()
print(results_brand1.summary())
```

## **What is the regression I ran?**

\begin{equation}
Units1=Intercept+\alpha_1 Price 1+\alpha_2 Price 2+ \alpha_3 Price 3+ \alpha_4 Feat_1+\alpha_5 Feat_2+\alpha_6 Feat 3+\alpha_7 Disp1+\alpha_8 Disp2+\alpha_9 Disp3+Errors
\end{equation}

```python
# how to define log
Ln_y_brand1 = np.log(y_brand1) #This taking the log of sales

# do the semi-log estimation main difference is Ln_y_brand1 in the call,
# i.e., reg_ln_brand1 = sm.OLS(Ln_y_brand1,x_brand1) instead of reg_brand1 = sm.OLS(y_brand1,x_brand1).
reg_ln_brand1 = sm.OLS(Ln_y_brand1,x_brand1)#Rest of the code follow by replacing reg_brand1 by reg_ln_brand1
type(reg_ln_brand1)
results_ln_brand1 = reg_ln_brand1.fit()
print(results_ln_brand1.summary())
```

```python
# c) Log-Log Model for Brand 1: log(UNITS1) on log(Price1), log(Price2), log(Price3), log(Price4), FEAT, DISP
x_brand1_log = df[['const','log_Price1','log_Price2','log_Price3','log_Price4','FEAT1','FEAT2','FEAT3','FEAT4','DISP1','DISP2','DISP3','DISP4']]
reg_loglog_brand1 = sm.OLS(Ln_y_brand1, x_brand1_log)
results_loglog_brand1 = reg_loglog_brand1.fit()
print("=== Brand 1 - LOG-LOG MODEL ===")
print(results_loglog_brand1.summary())
```

```python
# Model selection for Brand 1: compare Adj. R-squared
print("Brand 1 Model Comparison:")
print(f"  Linear:   Adj. R² = {results_brand1.rsquared_adj:.4f}")
print(f"  Semi-Log: Adj. R² = {results_ln_brand1.rsquared_adj:.4f}")
print(f"  Log-Log:  Adj. R² = {results_loglog_brand1.rsquared_adj:.4f}")
best_brand1 = max([
    ('Linear', results_brand1.rsquared_adj),
    ('Semi-Log', results_ln_brand1.rsquared_adj),
    ('Log-Log', results_loglog_brand1.rsquared_adj)
], key=lambda x: x[1])
print(f"\nRetained model for Brand 1: {best_brand1[0]} (highest Adj. R²)")

# Set coefficients for elasticity (based on retained model)
if best_brand1[0] == 'Linear':
    res = results_brand1
    coef_brand1_price1, coef_brand1_price2 = res.params.get('Price1',0), res.params.get('Price2',0)
    coef_brand1_price3, coef_brand1_price4 = res.params.get('Price3',0), res.params.get('Price4',0)
    brand1_elasticity_formula = 'linear'
elif best_brand1[0] == 'Semi-Log':
    res = results_ln_brand1
    coef_brand1_price1, coef_brand1_price2 = res.params.get('Price1',0), res.params.get('Price2',0)
    coef_brand1_price3, coef_brand1_price4 = res.params.get('Price3',0), res.params.get('Price4',0)
    brand1_elasticity_formula = 'semilog'
else:
    res = results_loglog_brand1
    coef_brand1_price1 = res.params.get('log_Price1',0); coef_brand1_price2 = res.params.get('log_Price2',0)
    coef_brand1_price3 = res.params.get('log_Price3',0); coef_brand1_price4 = res.params.get('log_Price4',0)
    brand1_elasticity_formula = 'loglog'
```

# What does this output tell us?

### Model Selection

We want to use the semi-log model because it might provide a higher adjusted $R^2$.

Remark: the $R^2$ represents the percentage of variations of the DV (i.e., y), that is explained by the "model".

This tells us if this model is a better fit to capture the variance in our data while minimizing error. An adjusted $R^2$ .57 tells us that 57% of the variance in the dependent variable (i.e., units) is captured by "the model" (The rest of the variance is captured by the error term)

### Variable Selection

Of all the variables, only FEAT1 is statistically significant. We know this because it has a p-value < .05.

This means that at a $95\%$ condifdence level, we cannot reject the null hypothesis that parameter estimates for Feature 1 is zero, i.e., therefore we conclude it is significant.





### Variable Analysis

Our price (Price1) doesn't significanlty impact sales ***for the price range in our data set***. It is inelastic for this price range. We do not need to do price optimization for this price range ***we can set the price at the max for this range, every time we discount the product, we are losing money that we could have made***

The competitors prices (Price2 and Price3) do not affect our sales either (within the data range). If they change their price, we do not need to react.

These prior two points tell us that we have ***market power***

## Brand 4 Regressions

For Brand 4, we run three model specifications:
- **a) Linear**: UNITS4 on Price4, DISP4, FEAT4, and all decision variables of other brands (Price1-3, FEAT1-3, DISP1-3)
- **b) Semi-Log**: log(UNITS4) on Price4, DISP4, FEAT4, and all decision variables of other brands
- **c) Log-Log**: log(UNITS4) on log(Price4), log(Price1), log(Price2), log(Price3), DISP4, FEAT4, and all other brands' DISP and FEAT (no log)

```python
# a) Linear Model: UNITS4 on Price4, DISP4, FEAT4 + all other brands' decision variables
y_brand4 = df['UNITS4']
x_brand4 = df[['const','Price4','Price1','Price2','Price3','FEAT4','FEAT1','FEAT2','FEAT3','DISP4','DISP1','DISP2','DISP3']]
reg_brand4 = sm.OLS(y_brand4, x_brand4)
results_lin_brand4 = reg_brand4.fit()
print("=== Brand 4 - LINEAR MODEL ===")
print(results_lin_brand4.summary())
```

```python
# b) Semi-Log Model: log(UNITS4) on Price4, DISP4, FEAT4 + all other brands' decision variables
Ln_y_brand4 = np.log(y_brand4)
reg_ln_brand4 = sm.OLS(Ln_y_brand4, x_brand4)
results_semilog_brand4 = reg_ln_brand4.fit()
print("=== Brand 4 - SEMI-LOG MODEL ===")
print(results_semilog_brand4.summary())
```

```python
# c) Log-Log Model: log(UNITS4) on log(Price4), log(Price1), log(Price2), log(Price3), DISP4, FEAT4 + other brands' DISP/FEAT
# Note: ALL price variables are logged; DISP and FEAT are NOT logged (log_Price created in price cell)
x_brand4_log = df[['const','log_Price4','log_Price1','log_Price2','log_Price3','FEAT4','FEAT1','FEAT2','FEAT3','DISP4','DISP1','DISP2','DISP3']]
reg_loglog_brand4 = sm.OLS(Ln_y_brand4, x_brand4_log)
results_loglog_brand4 = reg_loglog_brand4.fit()
print("=== Brand 4 - LOG-LOG MODEL ===")
print(results_loglog_brand4.summary())
```

```python
# Model selection for Brand 4: compare Adj. R-squared
print("Brand 4 Model Comparison:")
print(f"  Linear:   Adj. R² = {results_lin_brand4.rsquared_adj:.4f}")
print(f"  Semi-Log: Adj. R² = {results_semilog_brand4.rsquared_adj:.4f}")
print(f"  Log-Log:  Adj. R² = {results_loglog_brand4.rsquared_adj:.4f}")
best_brand4 = max([
    ('Linear', results_lin_brand4.rsquared_adj),
    ('Semi-Log', results_semilog_brand4.rsquared_adj),
    ('Log-Log', results_loglog_brand4.rsquared_adj)
], key=lambda x: x[1])
print(f"\nRetained model for Brand 4: {best_brand4[0]} (highest Adj. R²)")

# Set coefficients for elasticity calculation based on retained model
if best_brand4[0] == 'Linear':
    res_b4 = results_lin_brand4
    coef_brand4_price1 = res_b4.params.get('Price1', 0)
    coef_brand4_price2 = res_b4.params.get('Price2', 0)
    coef_brand4_price3 = res_b4.params.get('Price3', 0)
    coef_brand4_price4 = res_b4.params.get('Price4', 0)
    brand4_elasticity_formula = 'linear'  # elasticity = coef * (P_avg / Q_avg)
elif best_brand4[0] == 'Semi-Log':
    res_b4 = results_semilog_brand4
    coef_brand4_price1 = res_b4.params.get('Price1', 0)
    coef_brand4_price2 = res_b4.params.get('Price2', 0)
    coef_brand4_price3 = res_b4.params.get('Price3', 0)
    coef_brand4_price4 = res_b4.params.get('Price4', 0)
    brand4_elasticity_formula = 'semilog'  # elasticity = coef * P_avg
else:  # Log-Log
    res_b4 = results_loglog_brand4
    coef_brand4_price1 = res_b4.params.get('log_Price1', 0)
    coef_brand4_price2 = res_b4.params.get('log_Price2', 0)
    coef_brand4_price3 = res_b4.params.get('log_Price3', 0)
    coef_brand4_price4 = res_b4.params.get('log_Price4', 0)
    brand4_elasticity_formula = 'loglog'  # elasticity = coef directly
```
